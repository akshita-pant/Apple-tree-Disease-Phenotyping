## Data Prepartion

 Separating Test and Train data (Based on Labels) into different Folders

We have given input images in single folder.
- ***For further model building process we need to separate this data into train and test folders.***
- ***Separate train data into 4 different folders based on label category.***

```python
class_labels= train.columns[1:]
print(class_labels)
```

```python
# Creating new Label column into train.csv
num = 0
train['label'] =0
for i in class_labels:
    train['label'] = train['label'] + train[i]*num
    num +=1
train.head(5)
```

```python
# Creating Function to fetch label of individual train image
import re
def get_label(img):
    if re.search('Train',img):
        img = img.split('.')[0] # Fecth image_id from image name
        # return label corresponding to image_id
        label = train.loc[train['image_id'] == img]["label"]
        return label
    elif re.search('Test',img):
        pass
```

```python
path__dir =r'C:\Users\akshi\Desktop\Apple-Tree-Diseases-Predication-using-Image-Classification-main'
```

```python
# Creating New folders to separate data
shutil.os.mkdir(path__dir+ r'\train')
shutil.os.mkdir(path__dir+ r'\train\healthy')
shutil.os.mkdir(path__dir+ r'\train\multiple_diseases')
shutil.os.mkdir(path__dir+ r'\train\rust')
shutil.os.mkdir(path__dir+ r'\train\scab')
shutil.os.mkdir(path__dir+ r'\test')
```

```python
def copy_image_to_its_folder():
    images = natsort.natsorted(os.listdir(dir_path))  # Sorting images in proper form
    for img in tqdm(images):
        label = get_label(img)  # Assuming you have defined the get_label function
        path = os.path.join(dir_path, img)  # Original path of the image

        if 'Train' in img:  # Checking if the word 'Train' is in the image name
            if (img.split("_")[1].split(".")[0]) and label.item() == 0:
                shutil.copy(path, path__dir + r'\train\healthy')
            elif (img.split("_")[1].split(".")[0]) and label.item() == 1:
                shutil.copy(path, path__dir + r'\train\multiple_diseases')
            elif (img.split("_")[1].split(".")[0]) and label.item() == 2:
                shutil.copy(path, path__dir + r'\train\rust')
            elif (img.split("_")[1].split(".")[0]) and label.item() == 3:
                shutil.copy(path, path__dir + r'\train\scab')
        elif 'Test' in img:
            shutil.copy(path, path__dir + r'\test')
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
datagen = ImageDataGenerator(rescale = 1/255,
                            shear_range = 0.2,
                            zoom_range = 0.2,
                            horizontal_flip = True,
                            vertical_flip = True,
                            validation_split = 0.2)

train_datagen = datagen.flow_from_directory(Train_dir,
                                target_size = (256,256),
                                batch_size =16,
                                class_mode = 'categorical',
                                subset = 'training')
val_datagen = datagen.flow_from_directory(Train_dir,
                                         target_size = (256,256),
                                    batch_size =16,
                                    class_mode = 'categorical',
                                    subset = 'validation')
```

**(This code sets up image data generators for training and validation. It performs data augmentation, resizes images to 256x256 pixels, and creates generators for batches of 16 images each. The training generator is configured for categorical classification, and 20% of the data is reserved for validation. These generators are commonly used for training neural networks.)**

## CNN Model Building

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow.keras.layers import Dense,Activation,Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D,Dropout,BatchNormalization
```

```python
model=Sequential()
model.add(Conv2D(64,(3,3),activation='relu',padding='same',input_shape=(256,256,3))) 
#Convolutional layer with 64 filters, each of size (3,3), using the ReLU activation function. 
model.add(MaxPooling2D(2,2))
#Max pooling layer with a pool size of (2,2)->reduces the spatial dimensions.
model.add(Conv2D(64,(3,3),activation='relu',padding='same'))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(64,(3,3),activation='relu',padding='same'))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(128,(3,3),activation='relu',padding='same'))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(128,(3,3),activation='relu',padding='same'))
model.add(MaxPooling2D(2,2))
model.add(Flatten())
model.add(Dense(64,activation='relu'))
#A fully connected layer with 64 neurons and ReLU activation.
model.add(Dense(4,activation='softmax'))
#The final output layer with 4 neuronsand a softmax acti-funct. 
model.summary()
```

```python
model.compile(optimizer = 'adam', loss='categorical_crossentropy', 
              metrics=['accuracy'])
```

- **Optimizer**:- Adam adjusts the learning rate during training.
- **Loss Function**:- Categorical Crossentropy measures the difference between predictions and actual values for multi-class classification.
- **Metrics**:- Accuracy is used to evaluate the model's performance during training.

```python
#model callbacks
checkpoint=ModelCheckpoint(r'apple_tree_diseases.h5',
                          monitor='val_loss',
                          mode='min',
                          save_best_only=True,
                          verbose=1)
#Saves the best model weights based on validation loss to 'apple_tree_diseases.h5'.
#Monitors validation loss and updates weights when improvements are observed.
earlystop=EarlyStopping(monitor='val_loss',
                       min_delta=0.001,
                       patience=10,
                       verbose=1,
                       restore_best_weights=True)
#Halts training if validation loss does not improve significantly.
#Monitors validation loss, stops after 10 epochs of no improvement, and restores the best weights.

callbacks=[checkpoint,earlystop]
```

```python
#train cnn
model_history = model.fit(train_datagen, validation_data=val_datagen, epochs = 27, callbacks =callbacks)
```

the training versus validation accuracy/loss graph helps you assess how well your machine learning model is learning from the training data and generalizing to new, unseen data

```python
import numpy as np
import matplotlib.pyplot as plt

def moving_average(data, window_size):
    return np.convolve(data, np.ones(window_size)/window_size, mode='valid')

# Assuming model_history is a Keras history object
epochs = range(1, len(model_history.history['loss']) + 1)

# Smoothen the loss curves
window_size = 3  # Adjust the window size as needed
smoothed_train_loss = moving_average(model_history.history['loss'], window_size)
smoothed_val_loss = moving_average(model_history.history['val_loss'], window_size)

# Plot the smoothed loss curves
plt.plot(epochs[:len(smoothed_train_loss)], smoothed_train_loss, 'g', label='Training Loss')
plt.plot(epochs[:len(smoothed_val_loss)], smoothed_val_loss, 'b', label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Evaluate CNN model
cnn_model = load_model('cnn_model.h5')
cnn_val_datagen = datagen.flow_from_directory(Train_dir,
                                             target_size=(256, 256),
                                             batch_size=16,
                                             class_mode='categorical',
                                             subset='validation',
                                             shuffle=False)

cnn_predictions = cnn_model.predict(cnn_val_datagen)
cnn_predictions = np.argmax(cnn_predictions, axis=1)
cnn_true_labels = cnn_val_datagen.classes
cnn_f1 = f1_score(cnn_true_labels, cnn_predictions, average='weighted')

```

```python
# ResNet Model
resnet_model = Sequential()
resnet_model.add(ResNet50(include_top=False, pooling='avg', weights='imagenet'))
# removing the fully connected layer( ImageNet classification task)
#global avg-pooling:reduce the spatial dimensions of the output
resnet_model.add(Dense(4, activation='softmax'))

resnet_model.layers[0].trainable = False
#only the weights of the Dense layer will be updated, while the weights of the pre-trained ResNet50 layers remain fixed.
resnet_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Model callbacks
resnet_checkpoint = ModelCheckpoint('resnet_model.h5',
                                    monitor='val_loss',
                                    mode='min',
                                    save_best_only=True,
                                    verbose=1)

resnet_earlystop = EarlyStopping(monitor='val_loss',
                                 min_delta=0.001,
                                 patience=10,
                                 verbose=1,
                                 restore_best_weights=True)

resnet_callbacks = [resnet_checkpoint, resnet_earlystop]
```

```python
# Train ResNet model
resnet_model_history = resnet_model.fit(train_datagen, validation_data=val_datagen, epochs=30, callbacks=resnet_callbacks)
```

```python
# Evaluate ResNet model
resnet_model = load_model('resnet_model.h5')
resnet_predictions = resnet_model.predict(cnn_val_datagen)
resnet_predictions = np.argmax(resnet_predictions, axis=1)
resnet_f1 = f1_score(cnn_true_labels, resnet_predictions, average='weighted')
```

```python
# InceptionNet Model
inception_model = Sequential()
inception_model.add(InceptionV3(include_top=False, pooling='avg', weights='imagenet'))
inception_model.add(Dense(4, activation='softmax'))
```

```python
# Allow some layers of InceptionV3 to be trainable
inception_model.layers[0].trainable = True

inception_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Model callbacks
inception_checkpoint = ModelCheckpoint('inception_model.h5',
                                       monitor='val_loss',
                                       mode='min',
                                       save_best_only=True,
                                       verbose=1)

inception_earlystop = EarlyStopping(monitor='val_loss',
                                    min_delta=0.001,
                                    patience=10,
                                    verbose=1,
                                    restore_best_weights=True)

inception_callbacks = [inception_checkpoint, inception_earlystop]
```

```python
 # Train InceptionNet model with fine-tuning
inception_model_history = inception_model.fit(train_datagen, validation_data=val_datagen, epochs=30, callbacks=inception_callbacks)
```

```python
# Evaluate InceptionNet model
inception_model = load_model('inception_model.h5')
inception_predictions = inception_model.predict(cnn_val_datagen)
inception_predictions = np.argmax(inception_predictions, axis=1)
inception_f1 = f1_score(cnn_true_labels, inception_predictions, average='weighted')
```
